```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# مشخص کردن نوع داده‌ها
# فرض کنید که ستون‌های C1 تا C5 نامینال و ستون‌های C12 تا C32 اوردینال هستند
nominal_columns = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10',
                   'C11', 'C13', 'C14', 'C15', 'C32']
ordinal_columns = ['C12', 'C16', 'C17', 'C18', 'C19', 'C20', 'C21', 'C22',
                   'C23', 'C24', 'C25', 'C26', 'C27', 'C28', 'C29', 'C30',
                   'C31']

# تبدیل داده‌های نامینال به عددی با استفاده از Label Encoding
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for col in nominal_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))  # تبدیل به رشته برای جلوگیری از خطا
    label_encoders[col] = le  # ذخیره انکودرها برای استفاده‌های بعدی



ordinal_encoder = OrdinalEncoder()
data[ordinal_columns] = ordinal_encoder.fit_transform(data[ordinal_columns])
```

```python
from sklearn.model_selection import train_test_split

# جداسازی داده‌ها به مجموعه آموزشی و تست
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)
train_data_imputed = imputer.fit_transform(train_data)
test_data_imputed = imputer.transform(test_data)
```

```python
train_data_imputed = pd.DataFrame(train_data_imputed, columns=train_data.columns)
test_data_imputed = pd.DataFrame(test_data_imputed, columns=test_data.columns)
```
